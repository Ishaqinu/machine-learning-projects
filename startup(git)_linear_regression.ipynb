```python
import numpy as np
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/arib168/data/main/50_Startups.csv')
df
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
import seaborn as sns
sns.regplot(x=df['R&D Spend'],y=y,color='red')
```

```python
sns.regplot(x=df['Administration'],y=y,color='blue')
```

```python
sns.regplot(x=df['Marketing Spend'],y=y,color='black')
```

```python
#one hot encoding
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
col_trans=make_column_transformer((OneHotEncoder(handle_unknown='ignore'),['State']),remainder='passthrough')
x=col_trans.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
df=pd.DataFrame({'actual value':y_test,'predicted value':y_pred,'error':y_test-y_pred})
df
```

```python

from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,r2_score
mae=mean_squared_error(y_test,y_pred)
per=mean_absolute_percentage_error(y_test,y_pred)
sq=mean_squared_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print(mae,per,sq,score)
```

```python
# r2 score
# R^2 score = 1-RSS/TSS
# R^2=coiefient determinent
# RSS=sum of square bias
# TSS=total sum of square bias
```
