```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('/content/car economy price.csv')
data.head(5)
```

```python
# get dummy encoding
dummy=pd.get_dummies(data[['Location','Fuel_Type','Transmission',"Owner_Type"]],drop_first=True)
dummy
```

```python
df=pd.concat([data,dummy],axis=1)
df
```

```python
df.drop(['Unnamed: 0','Name','Location','Fuel_Type','Transmission','Owner_Type','New_Price'],axis=1,inplace=True)
```

```python
df.isna().sum()
```

```python
# string repalce
df['Mileage']=df['Mileage'].str.replace('km/kg','')
df['Mileage']=df['Mileage'].str.replace('kmpl','')
df['Engine']=df['Engine'].str.replace('CC','')
df['Power']=df['Power'].str.replace('bhp','')
df
```

```python
#null value replace
df['Mileage']=df['Mileage'].str.replace('null','0')
df['Engine']=df['Engine'].str.replace('null','0')
df['Power']=df['Power'].str.replace('null','0')
df
```

```python
df.dtypes
```

```python
# datatype change
df['Engine']=df['Engine'].astype(float)
df["Mileage"]=df['Mileage'].astype(float)
df['Power']=df['Power'].astype(float)
```

```python
df.dtypes
```

```python
df.loc[df.Engine==0,'Engine']=np.NaN
df.loc[df.Power==0,'Power']=np.NaN
df.loc[df.Mileage==0,'Mileage']=np.NaN
df
```

```python
df.isna().sum()
```

```python
#filling of missing value
df['Mileage']=df['Mileage'].fillna(df['Mileage'].mean())
df['Power']=df['Power'].fillna(df['Power'].mean())
df['Engine']=df['Engine'].fillna(df['Engine'].mean())
df['Seats']=df['Seats'].fillna(df['Seats'].mode()[0])
df
```

```python
df.isna().sum()
```

```python
y=df['Price']
x=df.drop(['Price','Fuel_Type_Electric'],axis=1)
x.dtypes
```
